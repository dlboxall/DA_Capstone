The functions below will download the dataset into your browser:


```python
from pyodide.http import pyfetch

async def download(url, filename):
    response = await pyfetch(url)
    if response.status == 200:
        with open(filename, "wb") as f:
            f.write(await response.bytes())
```

Prepare a bar plot of top ten countries by participants

```python
import matplotlib.pyplot as plt
import textwrap

# Step 1: Get top 10 countries
top10_countries = df['Country'].value_counts().head(10)

# Step 2: Wrap country names for the legend
wrapped_labels = [textwrap.fill(country, 30) for country in top10_countries.index]

# Step 3: Bar plot without x-axis labels
plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(top10_countries)), top10_countries.values, color=plt.get_cmap('tab10').colors)

# Step 4: Legend inside top-right corner of plot
plt.legend(bars, wrapped_labels, title="Country", loc='upper right', fontsize='small')

# Add titles and y-axis label only
plt.title('Top 10 Countries by Number of Survey Participants')
plt.ylabel('Number of Participants')
plt.xticks([])  # Hide x-axis labels

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import textwrap

# Step 1: Get top 10 countries and values
top10_countries = df['Country'].value_counts().head(10)
wrapped_labels = [textwrap.fill(country, 30) for country in top10_countries.index]
colors = plt.get_cmap('tab10').colors

# Step 2: Create plot
plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(top10_countries)), top10_countries.values, color=colors)

# Step 3: Add legend inside plot (top right)
plt.legend(bars, wrapped_labels, title="Country", loc='upper right', fontsize='small')

# Step 4: Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height + 150,  # slightly above bar
             f'{int(height)}',
             ha='center', va='bottom', fontsize=9)

# Step 5: Clean layout
plt.title('Top 10 Countries by Number of Survey Participants')
plt.ylabel('Number of Participants')
plt.xticks([])  # Hide x-axis labels
plt.tight_layout()

plt.show()
```
