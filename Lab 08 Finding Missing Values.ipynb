## Tasks


<h2>1. Load the Dataset</h2>
<p>
We use the <code>pandas.read_csv()</code> function for reading CSV files. However, in this version of the lab, which operates on JupyterLite, the dataset needs to be downloaded to the interface using the provided code below.

The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
# Basic structure of the dataframe
print(df.info())

# Summary statistics for numerical columns
print(df.describe())

# Summary statistics for categorical columns
print(df.describe(include='object'))

# Check for missing values
print(df.isna().sum().sort_values(ascending=False).head(10))  # top 10 with most missing values

# View unique values for key categorical columns (optional)
print(df['EdLevel'].value_counts())

```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
# Count of missing values per column (in descending order)
missing_values = df.isnull().sum().sort_values(ascending=False)
print("Missing values per column:\n", missing_values[missing_values > 0])
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Heatmap of Missing Values")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
missing_employment = df['Employment'].isnull().sum()
print(f"Missing rows in 'Employment': {missing_employment}")
```

```python
missing_EdLevel=df['EdLevel'].isnull().sum()
print(f"Missing rows in 'EdLevel': {missing_EdLevel}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
most_frequent_edlevel = df['EdLevel'].mode()[0]
print(f"Most frequent EdLevel: {most_frequent_edlevel}")

```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Fill missing values in EdLevel with the mode
df['EdLevel'].fillna(most_frequent_edlevel, inplace=True)

# Optional: verify imputation worked
print("Missing values in 'EdLevel' after imputation:", df['EdLevel'].isnull().sum())
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
import matplotlib.pyplot as plt

# Get value counts for EdLevel after imputation
edlevel_counts = df['EdLevel'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
edlevel_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Education Level After Imputation')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import textwrap

# Count values for EdLevel
ed_counts = df['EdLevel'].value_counts()

# Wrap and space legend labels
wrapped_labels = ["\n".join(textwrap.wrap(label, width=30)) + "\n" for label in ed_counts.index]

# Create the bar plot
colors = plt.get_cmap('tab10').colors
bars = plt.bar(range(len(ed_counts)), ed_counts.values, color=colors[:len(ed_counts)])

# Customize plot
plt.title('Distribution of Education Level (EdLevel)')
plt.ylabel('Count')
plt.xticks([])  # Remove x-axis tick labels

# Add legend outside to the right
plt.legend(
    bars,
    wrapped_labels,
    title='Education Level',
    loc='center left',
    bbox_to_anchor=(1.01, 0.5),
    fontsize='small',
    title_fontsize='medium',
    frameon=True
)

plt.tight_layout()
plt.show()

```
