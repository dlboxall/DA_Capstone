### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
import matplotlib.pyplot as plt

top_databases = df['DatabaseWantToWorkWith'].value_counts().nlargest(5)
ax = top_databases.plot.pie(autopct='%1.1f%%', figsize=(6, 6), title='Top 5 Databases Respondents Want to Work With')
ax.set_ylabel('')  # This line removes the y-axis label
plt.show()


```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
# Top 5 developer types from 'DevType' column
top_dev_types = df['DevType'].value_counts().nlargest(5)
top_dev_types.plot.pie(autopct='%1.1f%%', figsize=(6, 6), title='Top 5 Developer Roles')
ax.set_ylabel('')  # This line removes the y-axis label
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here
# Top 5 operating systems from 'OpSysProfessional' column
top_os = df['OpSysProfessional use'].value_counts().nlargest(5)
top_os.plot.pie(autopct='%1.1f%%', figsize=(6, 6), title='Top 5 Professional Operating Systems')
```

```python
# Split multi-response entries and explode into individual OSes
os_series = df['OpSysProfessional use'].dropna().str.split(';').explode()
top_os = os_series.value_counts().nlargest(5)

# Plot the cleaned-up pie chart
ax = top_os.plot.pie(autopct='%1.1f%%', figsize=(6, 6), title='Top 5 Professional Operating Systems')
ax.set_ylabel('')
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here
# Split and flatten the LanguageHaveWorkedWith column
langs_series = df['LanguageHaveWorkedWith'].dropna().str.split(';').explode()
top_langs = langs_series.value_counts().nlargest(5)

# Pie chart
ax = top_langs.plot.pie(autopct='%1.1f%%', figsize=(6, 6), title='Top 5 Programming Languages Worked With')
ax.set_ylabel('')
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
##Write your code here
# Split and flatten the NEWCollabToolsHaveWorkedWith column
tools_series = df['NEWCollabToolsHaveWorkedWith'].dropna().str.split(';').explode()
top_tools = tools_series.value_counts().nlargest(5)

# Pie chart
ax = top_tools.plot.pie(autopct='%1.1f%%', figsize=(6, 6), title='Top 5 Collaboration Tools Used')
ax.set_ylabel('')
```
